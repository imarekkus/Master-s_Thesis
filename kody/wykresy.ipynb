```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#from plotnine import *
```

# Wykres 1

```python
# Importing file

fig_1 = pd.read_csv("https://raw.githubusercontent.com/imarekkus/Master-s_Thesis/main/dane/Scopus-1656-Analyze-Year.csv", sep=';')
fig_1
```

```python
# Cleaning data

# Getting rid off "" in second column
fig_1["Year; Documents"] = fig_1["Year; Documents"].str.replace('"',"")

# Splitting columns into two
fig_1[['Year', 'Documents']] =  fig_1['Year; Documents'].str.split(';', expand=True)

# Dropping useless column
fig_1 = fig_1.drop(['Year; Documents'], axis=1)

# Order years
fig_1 = fig_1.sort_values(by=['Year'])
```

```python
# Checking types of data
fig_1.dtypes
```

```python
# Changing types of data in columns
fig_1['Documents'] = pd.to_numeric(fig_1['Documents'])
fig_1['Year'] = pd.to_datetime(fig_1['Year'], format='%Y').dt.year
```

```python
fig_1
```

```python
# Plotting figure
plt.figure(figsize=(13,6))
plt.style.use('seaborn')
plt.plot(fig_1['Year'], fig_1['Documents'], '-o')
plt.xlabel('Rok')
plt.ylabel('Liczba dokument√≥w')
plt.show()
```
